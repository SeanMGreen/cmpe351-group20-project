```python
#What factors significantly influence the pricing of Airbnb listings in a given location?



import pandas as pd

df = pd.read_csv('listings.csv')
```

```python
#Checking for missing data 

#To see the number of missing values in each column
missing_count = df.isnull().sum()

#To see columns with at least one missing value
missing_columns = df.columns[df.isnull().any()]

print("Missing count is:\n", missing_count)
print("Missing columns is:\n", missing_columns)

#Some columns have a significant number of missing values, such as reviews_per_month which has 5258 missing entries.

```

```python
#Plot out values to see

import seaborn as sns
sns.heatmap(df.isnull(), cbar=False)

# several columns have a significant amount of missing data (e.g., host_response_time, host_response_rate, host_acceptance_rate), while others have none or very little (e.g., id, last_scraped).
```
