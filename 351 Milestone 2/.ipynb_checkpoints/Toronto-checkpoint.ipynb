```python
#What factors significantly influence the pricing of Airbnb listings in a given location?
import numpy as np
import pandas as pd

df = pd.read_csv('TorontoListings.csv')
```

```python
#Columns i belive that provide little value and removed as a result

#id	listing_url	scrape_id	last_scraped	source	name	description	neighborhood_overview	picture_url host_url

#host_name host_location host_about host_thumbnail_url	host_picture_url host_verifications host_neighbourhood


#neighbourhood (just says tor ON) neighbourhood_group_cleansed bathrooms (empty) bedrooms amenities (blank)

#calendar_updated calendar_last_scraped

print("Before:", df.columns.tolist())

print("\n")

columns_to_remove = [
    'id', 'listing_url', 'scrape_id', 'last_scraped', 'source', 'name',
    'description', 'neighborhood_overview', 'picture_url', 'host_url',
    'host_name', 'host_location', 'host_about', 'host_thumbnail_url',
    'host_picture_url', 'host_verifications', 'host_neighbourhood',
    'neighbourhood', 'neighbourhood_group_cleansed', 'bathrooms', 'bedrooms',
    'amenities', 'calendar_updated', 'calendar_last_scraped'
]

# Remove the listed columns
df.drop(columns=[col for col in columns_to_remove if col in df.columns], inplace=True, errors='ignore')
print("After:", df.columns.tolist())
```

```python
#Formatting: Ensure that numerical variables are in the correct format

df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)

# Convert percentage strings to floats
df['host_response_rate'] = df['host_response_rate'].str.rstrip('%').astype(float) / 100
df['host_acceptance_rate'] = df['host_acceptance_rate'].str.rstrip('%').astype(float) / 100

```

```python
#Parsing Text

import re

# Example DataFrame
data = {
    'bathrooms_text': ['1 private bath', '2 baths', '1.5 shared baths', '2.5 baths', None]
}

# Function to extract numerical values
def extract_number(text):
    if pd.isnull(text):
        return None
    match = re.search(r'(\d+\.?\d*)', text)
    return float(match.group(1)) if match else None

# Apply the function to the bathrooms_text column
df['bathrooms'] = df['bathrooms_text'].apply(extract_number)

print(df[['bathrooms_text', 'bathrooms']])
```

```python
#Checking for missing data 

#To see the number of missing values in each column
missing_count = df.isnull().sum()

#To see columns with at least one missing value
missing_columns = df.columns[df.isnull().any()]

print("Missing count is:\n", missing_count)
print("Missing columns is:\n", missing_columns)

#Some columns have a significant number of missing values






```

```python
#Plot out values to see

import seaborn as sns
sns.heatmap(df.isnull(), cbar=False)

# several columns have a significant amount of missing data (e.g., host_response_time, host_response_rate, host_acceptance_rate), while others have none or very little (e.g., id, last_scraped).
```

```python
# Imputing missing data for rates with the median and correctly reassigning the results
df['host_response_rate'] = df['host_response_rate'].fillna(df['host_response_rate'].median())
df['host_acceptance_rate'] = df['host_acceptance_rate'].fillna(df['host_acceptance_rate'].median())

# Converting 'host_since' to datetime and filling missing values with the mode
df['host_since'] = pd.to_datetime(df['host_since'])
host_since_mode = df['host_since'].mode()[0]
df['host_since'] = df['host_since'].fillna(host_since_mode)

# Filling missing 'host_response_time' with the mode
host_response_time_mode = df['host_response_time'].mode()[0]
df['host_response_time'] = df['host_response_time'].fillna(host_response_time_mode)

# Filling missing 'beds' with the mode
beds_mode = df['beds'].mode()[0]
df['beds'] = df['beds'].fillna(beds_mode)

# Filling missing 'bathrooms' with the mode
bathrooms_mode = df['bathrooms'].mode()[0]
df['bathrooms'] = df['bathrooms'].fillna(bathrooms_mode)

# Removing rows with missing 'price' values
df = df.dropna(subset=['price'])

# Imputing missing 'has_availability' with the mode
has_availability_mode = df['has_availability'].mode()[0]
df['has_availability'] = df['has_availability'].fillna(has_availability_mode)

# Imputing missing review scores with the mean
df['review_scores_rating'] = df['review_scores_rating'].fillna(df['review_scores_rating'].mean())
df['review_scores_accuracy'] = df['review_scores_accuracy'].fillna(df['review_scores_accuracy'].mean())
df['review_scores_cleanliness'] = df['review_scores_cleanliness'].fillna(df['review_scores_cleanliness'].mean())
df['review_scores_checkin'] = df['review_scores_checkin'].fillna(df['review_scores_checkin'].mean())
df['review_scores_communication'] = df['review_scores_communication'].fillna(df['review_scores_communication'].mean())
df['review_scores_location'] = df['review_scores_location'].fillna(df['review_scores_location'].mean())
df['review_scores_value'] = df['review_scores_value'].fillna(df['review_scores_value'].mean())

# Creating flags for unrated listings
df['is_unrated'] = df['first_review'].isna() | df['last_review'].isna()


# Filling missing 'host_is_superhost' with the mode
superhost_mode = df['host_is_superhost'].mode()[0]
df['host_is_superhost'] = df['host_is_superhost'].fillna(superhost_mode)

# Checking for missing data
missing_count = df.isnull().sum()
missing_columns = df.columns[df.isnull().any()]

print("Missing count is:\n", missing_count)
print("Missing columns is:\n", missing_columns)
```

```python
#Variable Types

#Categorical Variables:
#room_type property_type neighbourhood_cleansed bathrooms_text license license
# neighbourhood_cleansed (eg, Hillcrest Village) property_type (eg Private room in home) room_type (eg private room) license (eg STR-2009-FDSXPB)
#host_since	(eg 2018-03-17) host_response_time (eg within an hour) first_review (eg 2018-04-30)	last_review (eg 2024-01-07)

from scipy import stats
import re
from datetime import datetime


# List of original categorical features before one-hot encoding
categorical_features = ['room_type', 'property_type', 'neighbourhood_cleansed', 'license']

for feature in categorical_features:
    if feature in df.columns:  # Check if the original feature exists
        # Group the data by the categorical feature and collect prices into lists for each category
        grouped_prices = df.groupby(feature)['price'].apply(list)

        # Perform ANOVA
        anova_result = stats.f_oneway(*grouped_prices)

        print(f"ANOVA test result for {feature}: F-statistic = {anova_result.statistic}, P-value = {anova_result.pvalue}\n")



#P-values are extremely low, well below the common threshold of 0.05, 
#indicating strong evidence against the null hypothesis of equal group means. 
#This means you have statistically significant evidence that the room_type, property_type, neighbourhood_cleansed, 
#and license have an impact on the price, with room_type and property_type showing particularly strong effects based on the F-statistics.

# One-Hot Encoding for nominal categorical variables
categorical_features = ['room_type', 'property_type', 'neighbourhood_cleansed', 'license']
df = pd.get_dummies(df, columns=categorical_features, drop_first=True)

# Convert date variables to datetime and create new features
df['host_since'] = pd.to_datetime(df['host_since'])
df['first_review'] = pd.to_datetime(df['first_review'])
df['last_review'] = pd.to_datetime(df['last_review'])


latest_date_in_dataset = df['last_review'].max()
df['days_since_last_review'] = (latest_date_in_dataset - df['last_review']).dt.days

df['host_since_year'] = df['host_since'].dt.year

# Mapping ordinal variable 'host_response_time' to numeric categories
response_time_mapping = {
    'within an hour': 1,
    'within a few hours': 2,
    'within a day': 3,
    'a few days or more': 4
}
df['host_response_time'] = df['host_response_time'].map(response_time_mapping)

# Drop the original 'bathrooms_text' column if no longer needed
df.drop(columns=['bathrooms_text'], inplace=True)

# Ensure no steps introduced additional missing values
missing_value_counts = df.isnull().sum()
print(missing_value_counts[missing_value_counts > 0])





#Numerical Variables: latitude, longitude, price, number_of_reviews calculated_host_listings_count_entire_homes, calculated_host_listings_count_private_rooms, calculated_host_listings_count_shared_rooms
#host_response_rate host_acceptance_rate minimum_nights, maximum_nights price



#Boolean: host_is_superhost host_has_profile_pic host_identity_verified instant_bookable
boolean_columns = ['host_is_superhost', 'host_has_profile_pic', 'host_identity_verified', 'instant_bookable']

# Map 't' to 1 and 'f' to 0 for each boolean column
for column in boolean_columns:
    if column in df.columns:
        df[column] = df[column].map({'t': 1, 'f': 0}).astype(float)

print(df[boolean_columns].head())
```

```python
#Descriptive Statistics: 
#For numerical data, you can calculate measures of central tendency (mean, median) and dispersion (standard deviation, variance, range, quartiles).


numerical_variables = [
    'latitude', 'longitude', 'price', 'number_of_reviews',
    'calculated_host_listings_count_entire_homes', 'calculated_host_listings_count_private_rooms',
    'calculated_host_listings_count_shared_rooms', 'host_response_rate',
    'host_acceptance_rate', 'minimum_nights', 'maximum_nights', 'price'
]

# Extract the specified numerical columns
df_numerical = df[numerical_variables]

# Calculate measures of central tendency
central_tendency = df_numerical.agg(['mean', 'median'])

# Calculate measures of dispersion including min, max, std, var
dispersion = df_numerical.agg(['std', 'var', 'min', 'max'])

# Calculate quartiles
quartiles = df_numerical.quantile([0.25, 0.5, 0.75])

# Combine all statistics
# Ensure all parts are DataFrames and concatenate them
stats = pd.concat([central_tendency, dispersion, quartiles])

print(stats)
```

```python
#Visualization: Visualize the data using histograms for numerical variables and bar charts for categorical variables to see the distribution of the data.
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the aesthetic style of the plots
sns.set(style="whitegrid")

# Numerical Variables for Histograms
numerical_variables = [
    'latitude', 'longitude', 'price', 'number_of_reviews',
    'calculated_host_listings_count_entire_homes', 'calculated_host_listings_count_private_rooms',
    'calculated_host_listings_count_shared_rooms', 'host_response_rate',
    'host_acceptance_rate', 'minimum_nights', 'maximum_nights', 'price'
]
#Kernel Density Estimate (KDE). The KDE is a smooth, continuous curve that estimates the probability density function of the underlying variable


# Categorical Variables for Bar Charts
categorical_features = [
    'room_type', 'property_type', 'neighbourhood_cleansed', 'host_response_time'
]

# Plotting Bar Charts for Categorical Variables
for feature in categorical_features:
    if feature in df.columns:
        plt.figure(figsize=(10, 6))
        # Countplot automatically counts the frequency of unique values
        sns.countplot(y=df[feature], order = df[feature].value_counts().index)
        plt.title(f'Frequency of {feature}')
        plt.xlabel('Frequency')
        plt.ylabel(feature)
        plt.show()

# Setting the aesthetic style of the plots
sns.set(style="whitegrid")

# Additional Numerical Variables
additional_numerical_variables = ['bathrooms', 'days_since_last_review', 'host_since_year']

# Combined list of numerical variables for histograms
all_numerical_variables = numerical_variables + additional_numerical_variables

# Plotting Histograms for Numerical Variables
for feature in all_numerical_variables:
    if feature in df.columns:
        plt.figure(figsize=(10, 6))
        sns.histplot(df[feature], kde=True, bins=30)
        plt.title(f'Distribution of {feature}')
        plt.xlabel(feature)
        plt.ylabel('Frequency')
        plt.show()
```

```python
#Handling Outliers: Check for any outliers in your data which might affect the distribution and later analysis.

import seaborn as sns
import matplotlib.pyplot as plt

# Example for 'price' variable
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['price'])
plt.title('Price Distribution')
plt.show()

Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

# Defining bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identifying outliers
outliers = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]
print("Number of outliers:", len(outliers))

df_filtered = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
print(f"Original size: {len(df)}, After removing outliers: {len(df_filtered)}")

non_numeric_columns = df.select_dtypes(exclude=[np.number]).columns
print("Non-numeric columns:", non_numeric_columns)
```

```python
#Correlations
#Numerical-Numerical: Calculate Pearson's correlation coefficient for pairs of numerical variables to see if any are strongly correlated with price.

# Select only numeric columns for correlation analysis
numeric_df = df.select_dtypes(include=[np.number])

# Calculate Pearson's correlation coefficient for pairs of numerical variables
numerical_correlations = numeric_df.corr()['price'].sort_values(ascending=False)

print("Correlation of numerical variables with price:\n", numerical_correlations)

# Visualize the correlation matrix using seaborn's heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix of Numeric Variables')
plt.show()
```
